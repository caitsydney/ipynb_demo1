```python
#what are the most common words greg uses in his email?

file_in = open("out.txt","r")

greg_words_list = []

for line in file_in:
    #print line
    line = line.split(", ")
    for word in line:
        if word != 'greg' and word != 'u': #add this part in as needed
            greg_words_list.append(word)

#print greg_words_list

from statlib import stats

frequencies = stats.itemfreq(greg_words_list)
#print frequencies

frequencies.sort(key=lambda x: x[1]) #used stackoverflow for this ("sort tuple list on second value")
#print frequencies

frequencies.reverse() #used google search ("reverse list in python")
#print frequencies

freq = []
label = []
for item in frequencies:
    label.append(item[0])
    freq.append(item[1])
print freq[:10] #modify so that you just see the first 10!
print label[:10]

from pylab import *

#display this in a visual!
figure(str(f), figsize=(6,6))
ax = axes([0.1, 0.1, 0.8, 0.8])
explode=(0, 0.05, 0, 0)
#c = ['slategray','darkseagreen','rosybrown','mistyrose']
pie(freq[:8], labels=label[:8],autopct='%1.1f%%', shadow=True, startangle=90,) #colors=c)
title("The Many Emails of Greg the Wilson", bbox={'facecolor':'0.8', 'pad':10})
show()

#how often does greg re-use a given word? how repetitive is he with diction?
#remember, freq is a list of how many times he has said each word

print "Mean:",stats.mean(freq)
print "Standard Deviation:",stats.stdev(freq)
```
